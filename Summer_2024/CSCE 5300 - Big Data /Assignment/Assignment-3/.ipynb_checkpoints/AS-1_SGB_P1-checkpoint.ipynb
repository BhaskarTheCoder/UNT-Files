```python
# Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, roc_curve, auc
import xgboost as xgb
import shap
from scipy.stats import chi2_contingency

# Load the diabetes dataset
diabetes_data = pd.read_csv("diabetes_df.csv")

# Split features (X) and target (y)
features = diabetes_data.drop('Target', axis=1)
target = diabetes_data['Target']

# Split the data into training and testing sets (80% train, 20% test)
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

# Create and fit default XGBoost model
default_xgb = xgb.XGBClassifier(random_state=42)
default_xgb.fit(X_train, y_train)

# Predict using default model and calculate accuracy
default_predictions = default_xgb.predict(X_test)
default_accuracy = accuracy_score(y_test, default_predictions)
print(f"Default XGBoost model accuracy: {default_accuracy:.4f}")

# Create and fit optimized XGBoost model with hyperparameters
optimized_xgb = xgb.XGBClassifier(learning_rate=0.02, n_estimators=600, objective='binary:logistic', nthread=1, random_state=42)
optimized_xgb.fit(X_train, y_train)

# Predict using optimized model and calculate accuracy
optimized_predictions = optimized_xgb.predict(X_test)
optimized_accuracy = accuracy_score(y_test, optimized_predictions)
print(f"Optimized XGBoost model accuracy: {optimized_accuracy:.4f}")

# Generate SHAP values for feature importance
explainer = shap.TreeExplainer(optimized_xgb)
shap_values = explainer.shap_values(X_test)

# Calculate ROC curve and AUC
prediction_probabilities = optimized_xgb.predict_proba(X_test)[:, 1]
false_positive_rate, true_positive_rate, thresholds = roc_curve(y_test, prediction_probabilities)
roc_auc = auc(false_positive_rate, true_positive_rate)

# Plot ROC curve
plt.figure(figsize=(8, 6))
plt.plot(false_positive_rate, true_positive_rate, color='darkorange', lw=2, label=f'ROC curve (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc="lower right")
plt.show()

# Calculate p-value using chi-square test
observed_frequencies = pd.crosstab(y_test, optimized_predictions)
chi2, p_value, dof, expected = chi2_contingency(observed_frequencies)
print(f"P-value for model acceptance: {p_value:.6f}")

# Plot histogram of prediction probabilities (confidence intervals)
plt.figure(figsize=(8, 6))
plt.hist(prediction_probabilities, bins=20, edgecolor='black')
plt.title('Histogram of Prediction Probabilities')
plt.xlabel('Predicted Probability')
plt.ylabel('Frequency')
plt.show()

# Print a short report with observations
print("\nShort Report:")
print(f"1. Default XGBoost model accuracy: {default_accuracy:.4f}")
print(f"2. Optimized XGBoost model accuracy: {optimized_accuracy:.4f}")
print(f"3. AUC (Area Under the ROC Curve): {roc_auc:.4f}")
print(f"4. P-value for model acceptance: {p_value:.6f}")
print("5. Feature Importance: SHAP plots should be generated separately")
print("6. Confidence Intervals: Refer to the histogram of prediction probabilities")
```

```python
# Summary plot with custom color
plt.figure(figsize=(10, 6))
shap.summary_plot(shap_values, X_test, plot_type="bar", color='#1E88E5', show=False)
plt.title("Feature Importance")
plt.tight_layout()
plt.show()

# Detailed SHAP plot for top features with custom color scheme
plt.figure(figsize=(10, 6))
shap.summary_plot(shap_values, X_test, plot_type="dot", 
                  color_bar=True, 
                  color_bar_label='Feature value',
                  cmap=plt.get_cmap("viridis"),  # You can change this to other color maps
                  show=False)
plt.title("SHAP Values")
plt.tight_layout()
plt.show()
```
